```
import os
import pandas as pd 
from google.colab import drive
from IPython.display import display, HTML 
from sklearn.linear_model import LinearRegression
drive.mount('/content/drive')

```

```

# Importing the dataset
rawpath="/content/drive/My Drive/SEAS-AU/Semester6/ML/ML Project/ML Dataset/IPL_points/"
count=0
dataset=pd.DataFrame()
exclude=["2009_IPL_points_DC.csv", "2009_IPL_points_SRH.csv", "2016_IPL_points_RR.csv"]
for filename in os.listdir(rawpath):

  if filename in exclude:
    continue
  else:
    path=rawpath+filename
    data = pd.read_csv(path, usecols = ['Pts', 'Wkts', 'Dots', '4s', '6s', 'Catches', 'Stumpings'])  
    dataset=pd.concat([dataset, data])
    # print("File: ", filename)
    count += 1

# display(dataset.head())
# X
features=dataset[['Wkts', 'Dots', '4s', '6s', 'Catches', 'Stumpings']]
#y
points=dataset[['Pts']]

print("\nTotal files", count, "Data set collected") 

```

```
linreg = LinearRegression().fit(features, points)
linreg.score(features, points)
print("coefficients: ", linreg.coef_)
print("intercept: ", linreg.intercept_)
print("Model applied")        
```
